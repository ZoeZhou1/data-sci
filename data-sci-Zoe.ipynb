```python
# import numpy as np
import pandas as pd
from sklearn import linear_model as MLR
from sklearn.preprocessing import PolynomialFeatures as PR
from sklearn.metrics import mean_squared_error as MSE
```

```python
train = pd.read_csv('train.csv')
train
```

```python
# cut (Fair, Good, Very Good, Premium, Ideal)
train.cut.replace(['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], [1,2,3,4,5], True)

# colour, from J (worst) to D (best)
train.color.replace(['J', 'I', 'H', 'G', 'F', 'E', 'D'], [1,2,3,4,5,6,7], True)

# clarity (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
train.clarity.replace(['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], [1,2,3,4,5,6,7,8], True)

train
```

```python
X_values = train[['carat', 'cut', 'color', 'clarity', 'depth', 'table']].values
y_values = train['price'].values

multilinear = MLR.LinearRegression()
multilinear.fit(X_values, y_values)
print(multilinear.coef_)
y_predictions = multilinear.predict(X_values)
MSE(y_values, y_predictions)
```

```python
degree=3
poly_model = PR(degree=degree)
poly_X_values = poly_model.fit_transform(X_values)
poly_model.fit(poly_X_values, y_values)

poly_reg_model = MLR.LinearRegression()
poly_reg_model.fit(poly_X_values, y_values)

poly_y_pred = poly_reg_model.predict(poly_X_values)
MSE(y_values, poly_y_pred)
```
